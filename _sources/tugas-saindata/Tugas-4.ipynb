## Penugasan piplines Proyek Sain Data,Langkah-langkah pengerjaan nya sebagai berikut :
1.   Pertama siapkan data iris dengan mengambil data di postgreesql elephant dengan menggunakan script
2.   kemudian kita aktifkan mysql,prosgresql,mysql server untuk memisahkan data yg ada di iris
3.   setelah mengaktifkan mysql kemudian masuk ke phpMyAdmin untuk import file iris
4.   untuk di posgresql setelah aktif langsung kita import data iris 
5.   sama seperti mysql dan postgresql disini setelah mengaktifkan mysql server langsung kita import data iris 
6.   setelah kita sudah memisahkan data iris menjadi beberapa sumber,kita langsung masuk ke link https://cloud.orchest.io/ kita log in terlebih dahulu setelah selesai login di cloud orchest kita akan membuat piplines untuk menghubungkan langkah-langkah memperoleh akurasi yg diinginkan menggunakan metode decision tree dan SVM.
7.   setelah login di cloud orchest kita langsung gabungkan data iris dari beberapa sumber menjadi satu
8.   data telah terkumpul kita langsung klasifikasi data tersebut untuk mengetahui akurasi yg terbaik dari kedua metode.
9.   akurasi dari dua metode telah muncul,selesai
